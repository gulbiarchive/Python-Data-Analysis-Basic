```python
import pandas as pd
import numpy as np
mpg = pd.read_csv('mpg.csv')
mpg['total'] = (mpg['cty'] + mpg['hwy'])/ 2
mpg['total'].mean()
mpg['total'].describe()
mpg['total'].plot.hist()
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg['test'].value_counts() # 연비 합격 빈도표 만들기
content_test = mpg['test'].value_counts()

# 연비 합격 빈도 막대 그래프 제작
# 이름이 회전되어 있다. -> 기본값이 축 이름을 90도 회전하게 되어있음
# rot = 0 # 축 이름 수평
content_test.plot.bar(rot = 0)
```

```python
import pandas as pd
import numpy as np

mpg = pd.read_csv('mpg.csv')

mpg['total'] = (mpg['cty'] + mpg['hwy'])/ 2
mpg['total'].mean()

mpg['grade'] = np.where(mpg['total'] >= 30, 'A',
                        np.where(mpg['total'] >= 20, 'B', 'C'))
content_grade = mpg['grade'].value_counts().sort_index()
content_grade.plot.bar(rot=0)
```
